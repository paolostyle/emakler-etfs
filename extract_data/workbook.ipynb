```python
import pandas as pd
import camelot
```

```python
tables = camelot.read_pdf("../data/source.pdf", pages="2-end", flavor="stream")
```

```python
data = []

for table in tables:
    data.append(table.df[8:])

df = pd.concat(data)
df = df.rename(
    columns={
        0: "stock_curr",
        1: "market_name",
        2: "etf_name",
        3: "long_ticker",
        4: "bloomberg_ticker",
        5: "google_ticker",
        6: "isin",
        7: "instrument_type",
    }
).reset_index()
df[["stock_market", "currency"]] = df["stock_curr"].str.extract(r"(.+) \(([A-Z]{3})\)")
df[["ticker", "market_code"]] = df["bloomberg_ticker"].str.split(":", 1, expand=True)
df["domicile"] = df["isin"].str.slice(start=0, stop=2)
df = df[["stock_market", "currency", "etf_name", "ticker", "market_code", "isin", "domicile"]]

df
```
